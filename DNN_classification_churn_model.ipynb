# Artificial/ DNN

```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
dataset=pd.read_csv("/content/Churn_Modelling.csv")
dataset.head(2)
```

```python
#row number,customer_id and Surname is irrelavent.
#split data into independent and dependent variable
x=dataset.iloc[:,3:-1]
y=dataset.iloc[:,-1]
```

```python
x
```

```python
# important, do encoding using get_dummies(onehot)
x=pd.get_dummies(x, columns=['Geography','Gender'])
x
```

```python
# drop dummies, bcz these columns that gets created after 1 hot encoding will
#be able to cover up, if female is 1, by default male will be 0
x = x.drop(['Geography_France','Gender_Female'], axis=1)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_std = sc.fit_transform(x)
x_std
```

```python
#split data into training n testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_std, y, test_size=0.2, random_state=1)
```

#building deep neural network(multilayerperceptron)

```python
from tensorflow import keras
```

```python
#create api
dnn= tf.keras.models.Sequential()
#adding first hidden layer(randomly 6 neurons)
dnn.add(tf.keras.layers.Dense(units=6, activation= 'relu'))

#adding 2nd hiiden layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

#adding output
dnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

#train the model
dnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#train the model
dnn.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
#predict the test result
y_pred=dnn.predict(x_test)
```

```python
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy=accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
classification_report(y_test, y_pred)
```
